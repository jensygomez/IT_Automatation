Para `create_file.py`:

```python
#!/usr/bin/env python3

import os
import sys

"""
Este script crea un nuevo archivo si no existe y escribe un mensaje en él. 
Si el archivo ya existe, imprime un mensaje de error y sale con un código de salida 1.

Ejemplo de uso:
    $ ./create_file.py example
    $ echo $?
    0

    $ ./create_file.py example
    Error, the file example already exists!
    $ echo $?
    1
"""

filename = sys.argv[1]  # Obtiene el nombre del archivo del primer argumento

if not os.path.exists(filename):  # Verifica si el archivo no existe
    with open(filename, "w") as f:  # Crea el archivo y escribe en él
        f.write("New file created\n")
else:
    print("Error, the file {} already exists!".format(filename))  # Imprime un mensaje de error si el archivo ya existe
    sys.exit(1)  # Sale con un código de salida 1
```

Esta documentación explica qué hace el script `create_file.py`, cómo usarlo y qué comportamiento esperar en diferentes situaciones.
